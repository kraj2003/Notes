### Range and IQR

Range and interquartile range (IQR) both *measure the "spread"* in a data set. Looking at spread lets us see how much data varies. Range is a quick way to get an idea of spread. It takes longer to find the IQR, but it sometimes gives us more useful information about spread.

Range is the difference between maximum and minimum number in the data set. range gives us an idea that how much the data is spreaded.

Interquartile Range (IQR)
Interquartile range is the amount of spread in the middle 50% of a dataset.
In other words, it is the distance between the first quartile and the third quartile.

Standard deviation measures the spread of a data distribution. It measures the typical distance between each data point and the mean.
The formula we use for standard deviation depends on whether the data is being considered a population of its own, or the data is a sample representing a larger population.
If the data is being considered a population on its own, we divide by the number of data points,***N*** .
If the data is a sample from a larger population, we divide by one fewer than the number of data points in the sample, (***n-1***).

### Box Plot
A box and whisker plot—also called a box plot—displays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum.

Statisticians have developed many ways to identify what should and shouldn't be called an outlier.
A commonly used rule says that a data point is an outlier if it is more than ***1.5.IQR*** above the third quartile or below the first quartile. Said differently, low outliers are below ***Q1-1.5.IQR*** and high outliers are above ***Q3-1.5.IQR***.

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.array([1,3,5,7,2,9,10,13,15,50]) #1,2,3,5,7,9,10,13,15,50
Average=np.mean(data)
print("Mean of the Data :", Average)

Q1=(25/100)*(len(data)+1)
print("25th % of the data",Q1)
#after sorting the data the 2.75th element is 25th element i.e 3. 25% of the data lies under 3.
Q3=(75/100)*((len(data))+1)
print("75th % of the data",Q3)
# 75% of the data lies under the Q3
fifth_percentile=(50/100)*(len(data)+1)
print("50th % of the data",fifth_percentile)
IQR=Q3-Q1
print("Inter Quantile range :", IQR)

#ouliers are those that does not fall in lower bracket till higher bracket
Lower_brac=Q1-1.5*(IQR)
Higher_brac=Q3+1.5*(IQR)
print("Lower bracket :", Lower_brac)
print("Higher bracket :", Higher_brac)

#any data point that doesn't fall in lower or higher bracket is an outlier
for i in data:
    if i <-6.0 :
        print("These are the Lower Outliers :",i)
    if i> 18:
        print("These are the higher Outliers:", i)
        

#now let's represent it visually
plt.figure(figsize=(10,2))
plt.boxplot(data,vert=False)
plt.show()
```

## Type I and Type II error
Both type 1 and type 2 errors are mistakes made when testing a hypothesis. A type 1 error occurs when you wrongly reject the null hypothesis (i.e. you think you found a significant effect when there really isn't one). A type 2 error occurs when you wrongly fail to reject the null hypothesis (i.e. you miss a significant effect that is really there)


### Type I error
