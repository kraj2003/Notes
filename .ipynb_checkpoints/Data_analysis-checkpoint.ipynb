### Missing values

+ Missing data is a common problem in most real-world scientific datasets. While the best way for dealing with missing data will always be preventing the occurrence in the first place, it will still remain a problem. 
+ **Cause of Missing values** : Sometimes data is collected from sensors that fail to record or data collection is distributed across individuals and the merged data does not harmonize well.
+ There are a variety of ways for dealing with missing data, from more simplistic to very sophisticated, but a standard metric by which we measure utility will still be model performance.

1. **Flag missing values** : One strategy for accounting for missing values is to simply ignore them. This is usually not a good idea because you often have little insight into how the missing data influenced the results.

    +  a minimum, you may need to convert missing elements to a flag that can be ingested by the model. For example, when dealing with missing values in a column of categorical data, an option is to treat “missing” as one of the categories for the model to train on.
    + In such a scenario, it is important to follow the standard practice of “dummifying” the categorical variable, such as with Pandas’ get_dummies method.
     
    + umerical :Converting a numerical missing value to a flag to denote that it is missing is generally not good practice, because the choice of the value of the flag will have implications on how your model treats such data. It is usually best to use an imputation technique.
2. **Complete Case Analysis**: The opposite strategy from tracking all your missing values is to delete either rows (observations) or columns (features), from your training data. In the case of deleting rows, this is called complete case analysis, and is quite common.
    + Complete case analysis can lead to undesirable results, but the degree to which it does depends on the category of missingness.
    + If you plan to use a predictive model on the data that you have, you will still need a plan to account for missing values. Complete case models in training can yield additional problems in the future. Sometimes complete-case analysis is used as a first pass through the workflow.
    + To remove data that contains missing values, Pandas library has a built in method called dropna. Essentially, with the dropna method, you can choose to drop rows or columns that contain missing values like NaN. So you'll need to specify axis = 0 to drop the rows, or axis = 1 to drop the columns that contain the missing values.Deleting rows by filtering out rows of Pandas DataFrames that have missing data is straightforward:
    
    + he default behavior of the .dropna ethod is to return a new DataFrame containing only the complete rows.This method naturally has additional functionality, such as dropping columns rather than rows:
    
    + Whole columns should be dropped only if most entries in the column are empty.
    + Any data entry without target data cannot be used for prediction; therefore any row now without target data is not useful to us. thus we will drop the whole row

### Categories of Missing Values
<br>
The categories of "missingness" can have important implications for statistical bias and power. The three categories of missing data are:

1. **Missing completely at random or MCAR:**
    + When data are MCAR, missing cases are, on average, identical to non-missing cases, with respect to the feature matrix. Complete case analysis will reduce the power of the analysis, but will not affect bias.
    + When data is MCAR, there is absolutely no relationship between the data missing and any other values, observed or missing, within the dataset. In other words, those missing data points are a random subset of the data. 
    + There is nothing systematic going on that makes some data more likely to be missing than other.
2. **Missing at random or MAR:**
    + When data are MAR the missing data often have some dependence on measured values, and models can be used to help impute what the likely data would be. 
3. **Missing not at random or MNAR:**
    + In this case the missing data depend on unmeasured or unknown variables. There is no information available to account for the missingness.
    
**Note** Of these three categories, the best case scenario is that the data are MCAR. It should be noted that imputing values under the other two types of missingness can result in an increase in bias. Many strategies have been proposed for the management and replacement of missing values—a process known as *imputation*.

### Simple Imputation

Once you have decided to fill in missing values (and grappled with the implications of doing so) the simplest approach is to treat each column/feature separately and use some chosen value, such as the mean of the available data as the imputation value.



As always, the best choice for exactly how and whether to impute missing data will depend on the nature of the data at hand and the overall goals of the project. You may need to try a few different options and compare how your model performs on some hold out data. Of course, at this point in the data science workflow you haven’t built a model yet, so you would just want to make a note to reconsider your imputation methods once you get to the back-and-forth between the feature engineering and modeling phases of the project. This sort of back-and-forth also underlies the big picture approach to missing values taken with multiple imputation.

### Normal Distribution

+ The normal distribution is a statistical distribution that is commonly used in many fields to describe continuous data that tend to cluster around a mean value. It is also known as the Gaussian distribution or the bell curve, because its shape is symmetrical and resembles a bell.
+ In a normal distribution, the data is spread out around the mean value according to a particular pattern, with most of the data concentrated in the middle and fewer data points toward the tails of the distribution.
+ The distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ), which determine the location and the spread of the distribution, respectively.

<b>Box Plot</b>
   + A box plot, also known as a box and whisker plot, is a graph that can show outliers in a set of data:
   
<i>What a box plot shows?</i>
   + A box plot shows the distribution of data, including the minimum and maximum values, the median, and the interquartile range.

<br> 
<b>Identifying outliers through Z-Score</b>

+ A z-score can help identify outliers by measuring how far a data point is from the mean of a set of data. The z-score is calculated using a formula, and then a threshold is set to determine which data points are considered outliers.

+ A z-score of greater than 2 or less than -2 is considered unusual, while a z-score of greater than 3 or less than -3 is considered highly unusual. 
 
<br>
The approach you choose for handling outliers depends on the nature of your data and the goals of your analysis. It's important to consider how removing or transforming outliers may impact the integrity of your results.

```python
# Practical inplementation of identifying outliers

import matplotlib.pyplot as plt
import numpy as np

data=np.array([1,3,5,7,2,9,10,13,15,50]) #1,2,3,5,7,9,10,13,15,50
Average=np.mean(data)
print("Mean of the Data :", Average)

Q1=(25/100)*(len(data)+1)
print("25th % of the data",Q1)
#after sorting the data the 2.75th element is 25th element i.e 3. 25% of the data lies under 3.
Q3=(75/100)*((len(data))+1)
print("75th % of the data",Q3)
# 75% of the data lies under the Q3
fifth_percentile=(50/100)*(len(data)+1)
print("50th % of the data",fifth_percentile)
IQR=Q3-Q1
print("Inter Quantile range :", IQR)

#ouliers are those that does not fall in lower bracket till higher bracket
Lower_brac=Q1-1.5*(IQR)
Higher_brac=Q3+1.5*(IQR)
print("Lower bracket :", Lower_brac)
print("Higher bracket :", Higher_brac)

#any data point that doesn't fall in lower or higher bracket is an outlier
for i in data:
    if i <-6.0 :
        print("These are the Lower Outliers :",i)
    if i> 18:
        print("These are the higher Outliers:", i)
        

#now let's represent it visually
plt.figure(figsize=(10,2))
plt.boxplot(data,vert=False)
plt.show()
```

## Imbalance data

Imbalanced data pertains to datasets where the distribution of observations in the target class is uneven. In other words, one class label has a significantly higher number of observations, while the other has a notably lower count.

<b>Problem with Handling Imbalanced Data for Classification?</b>
   + Algorithms may get biased towards the majority class and thus tend to predict output as the majority class.
   + Minority class observations look like noise to the model and are ignored by the model.
   + Imbalanced dataset gives misleading accuracy score.
   
<b> Ways to handle Imbalanced Data for Classification</b>
   1. <b> Under sampling : </b> 
       + Removes samples from the majority class to make the class proportions more balanced. 
       + This technique is useful when the majority class has many similar samples or when the dataset is large.
       +  However, undersampling can lead to biased models if it results in a loss of information.
       + We can use Under sampling by importing *NearMiss* or *RandomUnderSampler* from *iblearn.under_sampling*,  which randomly removes instances.
       
   2. <b> Over Sampling : </b>
       + Adds new samples to the minority class to increase its representation in the dataset. 
       + This technique is useful when there are limited samples of the minority class or when the dataset is small. 
       + However, oversampling can lead to overfitting if the synthetic data created isn't representative of the real data.
       + We can use Over sampling by importing *RandomOverSampler* from *iblearn.Over_sampling*
       
**Note** : Both oversampling & undersampling are ways to infuse bias where you take more samples from one class than the other to neutralize the effect of the imbalance that is either already present in the data or likely to be developed if the samples were taken purely at random.

   3. <b>SMOTE : </b>
       + The Synthetic Minority Oversampling Technique (SMOTE) addresses imbalanced datasets by synthetically generating new instances for the minority class. 
       + Unlike simply duplicating records, SMOTE enhances diversity by creating artificial instances.
       + In simpler terms, SMOTE examines instances in the minority class, selects a random nearest neighbor using k-nearest neighbors, and generates a synthetic instance randomly within the feature space.

```python
# JOINNING TWO DATAFRAMES
import pandas as pd
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)


frames = [df1, df2]

result = pd.concat(frames)

result
```

###  Join
is used to combine dataframes based on a common column. The dataframes must have at least one column in common, and the rows in the resulting dataframe will be the rows from the first dataframe that have matching values in the common column to the rows in the second dataframe.

### Merge
is similar to join, but it is more flexible. Merge can be used to combine dataframes based on multiple columns, and it can also be used to perform different types of joins, such as inner joins, outer joins, and left joins.

# Outliers

Outliers are data points that significantly deviate from the majority of the data. They can be caused by errors, anomalies, or simply rare events.

why are outliers problematic
Outliers can negatively impact the performance of machine learning models in several ways:

1. Overfitting: Models can focus on fitting the outliers rather than the underlying patterns in the majority of the data.
2. Reduced accuracy: Outliers can pull the model’s predictions towards themselves, leading to inaccurate predictions for other data points.
3. Unstable models: The presence of outliers can make the model’s predictions sensitive to small changes in the data.

how to detect outliers
There are several methods for detecting outliers, including:

1. Distance-based measures: These measures, like Z-score and interquartile range (IQR), calculate the distance of a data point from the center of the data distribution.
2. Visualization techniques: Techniques like boxplots and scatter plots can visually identify data points that lie far away from the majority of the data.
3. Clustering algorithms: Clustering algorithms can automatically group similar data points, isolating outliers as separate clusters.

```python
# 1. distance-based measure
#  Most observations in any dataset have the Z-score between -3 and 3. Any observation with the Z-score less than -3 or more 
# than 3 is considered to be an outlier as it is very unlikely
import numpy as np
df = pd.DataFrame(np.random.randint(100, 200, size=(5, 3)), columns=['A', 'B', 'C'])
df
from scipy.stats import zscore
df.apply(zscore)
```

how to handle the outliers
There are several approaches to handling outliers in machine learning:

1. Removing outliers: This is a simple approach but can lead to information loss.
2. Clipping: Outliers are capped to a certain value instead of being removed completely.
3. Transformation: Data can be transformed to reduce the impact of outliers, such as using log transformations for skewed data.
4. Robust models: Certain models are less sensitive to outliers, such as decision trees and support vector machines.

when should we remove the outliers
Removing outliers can be beneficial when they are likely due to errors or anomalies. However, it should be avoided when outliers represent genuine, albeit rare, occurrences within the data.
